```python
# Question 1.4-(a)
import gurobipy as gp
from gurobipy import GRB
import numpy as np

try:
# Create a new model
    PwC = gp.Model ("PwC-seqIP")

# processing times of jobs
    protime = [12, 23, 34, 46,25,8,5]

# # of machines
    m = 3

# range of jobs and machines
    jobs = range(len(protime))
    machines = range(m)

# Create variables
    x = PwC.addVars(machines, jobs, vtype = GRB.BINARY, name = "assign")
    C = PwC.addVars(machines, vtype = GRB.CONTINUOUS, name = "completion")
    Cmax=PwC.addVar(vtype=GRB.CONTINUOUS, name="Maximum Span")

# Set objective
    PwC.setObjective(Cmax, GRB.MINIMIZE)

# Add constraints
    PwC.addConstrs((gp.quicksum(x[i,j] for i in machines) == 1 for j in jobs))
    PwC.addConstrs((Cmax >=C[i] for i in machines))
    PwC.addConstrs((C[i] == gp.quicksum(protime[k] * x[i,k] for k in jobs) for i in machines))

  
    
# Optimize model
    PwC.optimize()
    PwC.write("PwC-seqIP.lp")

# print optimal solutions
    for v in PwC.getVars ():
        print ('%s %g' % (v.varName , v.x))

except gp.GurobiError as e:
    print ('Error code' + str (e. errno ) + ': ' + str(e))
except AttributeError :
    print ('Encountered an attribute error')
```
